```python
import pandas as pd
from elasticsearch import Elasticsearch
import logging
```

```python
# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


def post_elastic():
    es = Elasticsearch([{'host': 'localhost', 'port': 9200}])

    if not es.ping():
        logger.error("Gagal terhubung ke Elasticsearch")
        return
    else:
        logger.info("Koneksi ke Elasticsearch berhasil")

    df_clean = pd.read_csv('F:\Harun\Hacktiv8\Phase2\week2\project-m3\dags\P2M3_Harun_data_clean.csv')

    for i, row in df_clean.iterrows():
        doc = row.to_dict()
        es.index(index="store", id=i + 1, body=doc)  # ID yang unik

    logger.info("Data berhasil dikirim ke Elasticsearch")
```

```python
post_elastic()
```
